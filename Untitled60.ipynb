```python
import pylab

def binomialCoeff(n, k):
    result = 1
    for i in range(1, k+1):
        result = result * (n-i+1) / i
    return result

def Z(N, L, sigma):
    freespace = L - 2.0 * N * sigma
    if freespace > 0.0:
        result = freespace ** N
    else:
        result = 0.0
    return result

def pi(x, N, L, sigma):
    tot = 0.
    for k in range(0, N):
        Z1 = Z(k, x - sigma, sigma)
        Z2 = Z(N - k - 1, L - x - sigma, sigma)
        tot += binomialCoeff( N - 1, k) * Z1 * Z2
    Ztotal = Z(N, L, sigma)
    return tot / Ztotal

L = 20.0
N = 10
sigma = 0.75
xr = pylab.linspace(0.0, L, 201)
yr = [pi(x, N, L, sigma) for x in xr]
pylab.plot(xr, yr, 'red', linewidth=2.0)
pylab.xlabel('$x$', fontsize=14)
pylab.ylabel('$\pi(x)$', fontsize=14)
pylab.title('Exact density of %i clothes-pins ($\sigma$=%s)\non a line of length L=%s' % (N, sigma, L))
pylab.savefig('plot-direct_pins_density.png')
pylab.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def binomialCoeff(n, k):
    result = 1
    for i in range(1, k+1):
        result = result * (n-i+1) / i
    return result

def Z(N, L, sigma):
    freespace = L - 2.0 * N * sigma
    if freespace > 0.0:
        result = freespace ** N
    else:
        result = 0.0
    return result

def pi(x, N, L, sigma):
    tot = 0.
    for k in range(0, N):
        Z1 = Z(k, x - sigma, sigma)
        Z2 = Z(N - k - 1, L - x - sigma, sigma)
        tot += binomialCoeff(N - 1, k) * Z1 * Z2
    Ztotal = Z(N, L, sigma)
    return tot / Ztotal

# Parameters
L = 20.0
N = 10
sigma = 0.75

# Generate x values
xr = np.linspace(0.0, L, 201)

# Compute y values (probability density)
yr = [pi(x, N, L, sigma) for x in xr]

# Plotting
plt.plot(xr, yr, 'red', linewidth=2.0)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$\pi(x)$', fontsize=14)
plt.title('Exact density of %i clothes-pins ($\sigma$=%s)\non a line of length L=%s' % (N, sigma, L))
plt.savefig('plot-direct_pins_density.png')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def roll_die(num_rolls):
    """Simulate rolling a die and return the frequency of each outcome."""
    results = [random.randint(1, 6) for _ in range(num_rolls)]
    frequency = {outcome: results.count(outcome) for outcome in range(1, 7)}
    return frequency

def plot_distribution(frequency, num_rolls):
    """Plot the frequency distribution of die rolls."""
    outcomes = list(frequency.keys())
    counts = list(frequency.values())

    plt.bar(outcomes, counts, edgecolor='black')
    plt.xticks(outcomes)
    plt.xlabel('Outcome')
    plt.ylabel('Frequency')
    plt.title(f'Die Roll Distribution ({num_rolls} Rolls)')
    plt.show()

# Simulate and plot for 10, 100, and 10,000 rolls
for num_rolls in [10, 100, 10000]:
    frequency = roll_die(num_rolls)
    print(f"\nResults for {num_rolls} rolls:")
    for outcome, count in frequency.items():
        print(f"Outcome {outcome}: {count} times")
    plot_distribution(frequency, num_rolls)
```
